Построение регрессионных моделей с помощью библиотекии scipy.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * (x ** -2) + b * np.exp(x * c)

xdata = np.linspace(0.1, 4, 50)
y = func(xdata, 2.5, 5., 1.2)
rng = np.random.default_rng()
y_noise = 10 * rng.normal(size=xdata.size)
ydata = y + y_noise
plt.scatter(xdata, ydata, s=2, c='b', label='data')

popt, pcov = curve_fit(func, xdata, ydata)
plt.plot(xdata, func(xdata, *popt), 'r-',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

print(func(2, *popt))
```
